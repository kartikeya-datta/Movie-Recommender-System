I have to modify the format of the genres to a list formant.

Such as ['Action', 'Adventure', 'Fantasy', Science Fiction']


```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = [sublist[:3] for sublist in movies['cast'].apply(convert)]  # I am taking only first 3 actors, who would value the most in a movie
```

```python
movies['cast']
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director']) # I am taking only the director of the movie
movies['crew']
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split()) # I am splitting the overview into words and making it a list.append
movies['overview']
```

```python
movies['tagline'] = movies['tagline'].apply(lambda x: x.split() if isinstance(x, str) else [])
movies['tagline']
```

```python
movies.head()
```

Now I would like to remove spaces so that the recommender system will have one entiy for a name of a person or so. For example 'Sam Worthington, is a single entity and 'Sam Mendes' is a single entity but if i make tags with them now I will get three different tag with 'Sam' 'Worthington' and 'Mendes' which would desrupt the algorithms search mechanism. rather It would be efficient and porper to search for a single word 'SamWorthington' or 'SamMendes' as a single entity.


```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['tagline'] +movies['cast'] + movies['crew']
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x)) # I am joining all the words in the list to make a single string
movies['tags'] = movies['tags'].apply(lambda x: x.lower()) # I am converting all the words to lower case
```

```python
movies_df = movies[['movie_id', 'title', 'tags']].copy()
movies_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_df['tags'] = movies_df['tags'].apply(stem)
```

We need to find the similarities in the Tags data as they are in string format. I need to convert it into numerical representation to compare and predict the similarities between the movies to recommend them. Hence I am going to do Text vectorization of the data using **Bag of Words**. But whiel vectorizing the data we will exclude the stop wrods from our vectorization to improve our models performance as they do not make any impact on our movies similarity.


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()# I am converting the words into a matrix of token counts (a numpy array)
vectors
```

```python
cv.get_feature_names_out()
```
